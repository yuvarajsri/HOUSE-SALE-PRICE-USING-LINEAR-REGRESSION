VERFYING THE MULTICOLINEARITY AMONG INDENPENDENT VARAIABLES

```
C=X.corr()
C
v=[[str(i),str(j)]for i in C.columns for j in C.columns if(C.loc[i,j]>abs(0.5))&(i!=j) ]
```

```
v ,len(v)
```

```
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=X
```

```
vif_of_data=pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
```

```
vif_of_data
```

```
def MC_remover(data):
  vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index = data.columns)
  if vif.max() > 5:
    print(vif[vif == vif.max()].index[0],'has been removed')
    data = data.drop(columns = [vif[vif == vif.max()].index[0]])
    return data
  else:
    print('No Multicollinearity present anymore')
    return data
```

```
for i in range(7):
  vif_data = MC_remover(vif_data)

vif_data.head()
```

```
vif_of_data=pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
```

```
vif_of_data
```

IMPLEMENTING LINEAR REGRESSION

```
X = vif_data
Y = data['Sale_Price']
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 1)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```
model.fit(x_train,y_train)
```

```
y_pred=model.predict(x_test)
```
